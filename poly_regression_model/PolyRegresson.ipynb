## importing libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### importing dataset

```python
dataset = pd.read_csv('Position_Salaries.csv')
dataset
```

```python
X = dataset.iloc[:,1:-1].values
y = dataset.iloc[:,-1].values
```

### Training the linear regression model on the whole dataset

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X,y)
```

#### Training the PolyRgression model on the whole dataset

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_reg = PolynomialFeatures(degree=2)
X_poly = poly_reg.fit_transform(X)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly,y)
```

### visualising the linear regression results

```python
plt.scatter(X,y, color='red')
plt.plot(X, lin_reg.predict(X), color='blue')
plt.title('Truth or bluf (linear Regression)')
plt.xlabel('Postion label')
plt.ylabel('Salaries')
plt.show()
```

##visualsing ploy regression results

```python
plt.scatter(X,y, color='red')
plt.plot(X, lin_reg_2.predict(X_poly), color='blue')
plt.title('Truth or bluf (Poly Regression)')
plt.xlabel('Postion label')
plt.ylabel('Salaries')
plt.show()
```

## visualing poly regression model with higher resolution

```python
poly_reg = PolynomialFeatures(degree=4)
X_poly = poly_reg.fit_transform(X)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly,y)
```

```python
plt.scatter(X,y, color='red')
plt.plot(X, lin_reg_2.predict(X_poly), color='blue')
plt.title('Truth or bluf (Poly Regression)')
plt.xlabel('Postion label')
plt.ylabel('Salaries')
plt.show()
```

### predicting the salary at level 6.5 using Linear regression model

```python
lin_reg.predict([[6.5]])
```

##### predicting the salary at level 6.5 using poly regression model

```python
lin_reg_2.predict(poly_reg.fit_transform([[6.5]]))
```
